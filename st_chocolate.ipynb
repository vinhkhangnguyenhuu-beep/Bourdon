## Filtering Data

We can use `df[df[column_name] == 'Value']` to filter data. Now try filtering rows to show `"Sales Person"` whose name is `"Karlen McCaffrey"`.

```python
#print country starting with u
for item in my_csv_df['Country']:
    if item.startswith('U'):
        print(f'{item} starts with U')
```

If this syntax looks confusing, don't panic! All we're doing is putting the filter condition inside `df[]`, and we are simply keeping the rows where that condition is true. 

```python
#Karlen repeats multiple times.
#I want to extract only the rows with his name

#One shot way
#1: get into Dataframe.
#which column has name
my_csv_df['Sales Person'] == 'Karlen McCaffrey'
#get false and trues,
#but i want the rows, with data, not booleans

#so:
my_csv_df[my_csv_df['Sales Person'] == 'Karlen McCaffrey']
#double bracket returns dataframe

#now extract data of
#can store
new_df = my_csv_df[my_csv_df['Sales Person'] == 'Karlen McCaffrey']

#why twice mention? make dataframe from list?
```

```python
#other method

person_condition_filter = my_csv_df['Sales Person'] == 'Karlen McCaffrey'

my_csv_df[person_condition_filter]

#make a new dataframe with list and make that list into a dataframe
```

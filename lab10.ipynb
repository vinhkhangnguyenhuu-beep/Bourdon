3. Write a function called `rotate` that takes a list of items and shifts each element to the right by one cell.  The last element will be shifted to index 0.  You may not create any additional containers such as lists and dictionaries.  You must use while loops and your function should return the updated list.  

```python
def rotate (lst):
    last = lst[-1]
    i = len(lst)-2
    
    while i>=0:
        lst[i+1]=lst[i]
        i=i-1
    lst[0] = last
    return lst

listToTest = [1,2,3,4,5,6,7,8,9]
print (rotate(listToTest))
```

4. Write a function called `convert` that takes a list of items from which creates a dictionary and returns that dictionary.  The dictionary keys are the items of the list and the values are the number of times each item appears in the list.  

```python
def convert(listOfItems):
    CreatedDictionary = {}
    for item in listOfItems:
        itemCount = listOfItems.count(item)
        CreatedDictionary[item] = itemCount
    return CreatedDictionary

Lst = ['James', 'James', 'James', 'James', 'eats', 'breakfast', 'breafkast']
print (convert(Lst))
```

5. Write a function called `convert2` that takes a list of items from which creates a dictionary and returns that dictionary. Note that the list of items may contain any type of items such as strings, ints, floats, booleans.   

```python
#to redo 

def convert2 (list_items):
    Creation = {}
    for key in list_items:
        if len(str(key)) == 4:
            if Creation.get(key) == None:
                Creation.update({key: 1})

            else: 
                Creation.update({key: Creation.get(key)+1})
        

    return Creation
```

8.  Write a function called `divisors` that takes a positive integer and returns a list of all of its positive divisors.  

```python
#redo

def divisors(posInt):
    divisor = 1
    listDivisors = []
    while divisor <= posInt:
        if posInt % divisor ==0:
            listDivisors.append(divisor)
        divisor += 1
    return listDivisors
```

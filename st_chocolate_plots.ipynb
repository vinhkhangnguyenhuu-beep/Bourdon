## Line graph

Create a line graph using `plt.plot()` that:

```python
#figure out y points
ypoints = age_df['age']
#pull the column that represents the ypotns
plt.plot(ypoints, marker = 'o', label = 'age') 
#x points are the indeces if not provided
#I want each ypoints to be a o
plt.xlabel('Indeces')
plt.ylabel('Age')
plt.title('age Line Graph')
plt.legend()
# i want gridlines
plt.grid(True)
#how does it know we need to put age? label links to legend
plt.show()

#shows the zero index, the one
#plt.show() is always the last thing. Print of plots
```

## Bar chart

Create a bar chart using `plt.bar()` that:
1. shows the age for each name
2. make your bar chart green 

- label x-axis with *name*
- label y-axis with *age*

```python
ypoints = age_df['age']
xpoints = age_df['name']
#i need to provide the xpoints of the xaxis now
plt.bar (xpoints, ypoints, color = 'Brown') 
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('age Line Graph')
plt.xticks(rotation = 45)
plt.show()

#I want to rotate the x ticks

#plt.plot = line
#graded lab 3: just copy paste, change names, 
#default color is blue
```

## Pie chart and Save figure function

Create a pie chart using `plt.pie()` that:
1. shows the age for a given name.  **Hint:** use the name as labels


- add a legend
- call `show()` function from matplotlib library by writing `plt.show()`
- finally save the chart by calling `plt.savefig()`

```python
ypoints = age_df['age']
labels = age_df['name']
#no xpoints on pie chart
#whenever we have labels, have legend
plt.pie (ypoints, labels = labels) 
plt.title('age Line Graph')
plt.legend(loc = 'lower left')
plt.savefig('pieChart.png')
plt.show()


#if legend overlaps
#loc =              location of the legend
#can save picture or file


#write to directly save, save before the show
```
